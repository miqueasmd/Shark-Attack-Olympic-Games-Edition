##### Cleaning hour column:

```python
def get_hour(x):
    if isinstance(x, str):
        return x[0:2]
    return None

df["hour"]=df["time"].apply(lambda x: get_hour(x))
```

```python
def categorize_hour(hour):
    if isinstance(hour, str) and hour.isdigit():
        hour = int(hour)
    elif isinstance(hour, str):
        try:
            hour = int(hour.split("h")[0])
        except (ValueError, IndexError):
            return None

    if isinstance(hour, (int, float)):
        if 6 <= hour < 14:
            return "Morning"
        elif 14 <= hour < 19:
            return "Afternoon"
        elif 19 <= hour < 24:
            return "Evening"
        else:
            return "Night"
    elif hour in ['Mo', 'Af', 'Ni', 'La', 'Ev']:
        if hour == 'Mo':
            return "Morning"
        elif hour == 'Af':
            return "Afternoon"
        elif hour == 'Ni':
            return "Night"
        elif hour == 'La':
            return "Night"
        elif hour == 'Ev':
            return "Evening"
    return None


df["hour_category"]=df["hour"].apply(lambda x: categorize_hour(x))
```

Cleaning activity

```python
def clean_activity(activity):
    if pd.isna(activity):  # Check for NaN values
        return 'other activity'
    activity = activity.lower()  # Convert to lowercase for case-insensitive matching

    # Define the replacement rules
    if 'swimming' in activity or 'swim' in activity:
        return 'swimming'
    elif 'diving' in activity or 'dive' in activity:
        return 'diving'
    elif 'surfing' in activity or 'surf' in activity:
        return 'surfing'
    elif 'canoe' in activity:
        return 'canoe'
    elif 'kayak' in activity or 'kayaking' in activity:
        return 'kayaking'
    elif 'sail' in activity or 'sailing' in activity:
        return 'sailing'
    elif 'row' in activity or 'rowing' in activity:
        return 'rowing'
    elif 'kiteboard' in activity or 'kiteboarding' in activity or 'kite boarding' in activity:
        return 'kiteboarding'
    elif 'kite surf' in activity or 'kitesurf' in activity or 'kite surfing' in activity or 'kitesurfing' in activity:
        return 'kitesurfing'
    else:
        return 'other activity'

# Apply the cleaning function to the activity column
df['activity'] = df['activity'].apply(clean_activity)

# Show the cleaned dataset
print(df['activity'].value_counts())
```

```python
# Filter the DataFrame to keep only fatal accidents
fatal_accidents = df_usa[df_usa['injury'].str.contains('fatal', case=False, na=False)]
# fatal_accidents.value_counts().sort_values(ascending=False)

# Group by 'species' and count the number of fatal accidents
fatal_accidents_by_species = fatal_accidents.groupby('species_').size().reset_index(name='count')
fatal_accidents_by_species = fatal_accidents_by_species.sort_values(by='count', ascending=False)
# Display the grouped data
print(fatal_accidents_by_species)

# Assuming fatal_accidents_by_species is already defined
# Sort by the count and select the top 5 species
top_5_species = fatal_accidents_by_species.nlargest(5, 'count')

# Show pie chart of fatal accidents by species
plt.figure(figsize=(10, 6))
top_5_species.set_index('species_')['count'].plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribution of Fatal Shark Attacks by Top 5 Species')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
